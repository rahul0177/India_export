# Export 2010-2018

# This dataset includes the trade data for India for commodities 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("export.csv")
df.head(20)
```

# Here I Filter Out all items which are Highly Exported from India and also which are less Exported from India.

```python
max_export_value=df["value"].max()
max_export_row=df[df["value"]==max_export_value]
print("Highly Export item from india :-\n")
print(max_export_row[["Commodity","value"]].head())
print()
print()
min_export_value=df["value"].min()
min_export_row=df[df["value"]==min_export_value]
print("least Export item from india :-\n")
print(min_export_row[["Commodity","value"]].head(20))
```

```python
plt.hist(df.value,bins=[1,200,400,600,800,1000],rwidth=0.90,color=["green"])
plt.xlabel("Item Exported in Tone")
plt.ylabel("Items Count")
plt.title("Indian export Range")
plt.show()
```

so from here it is cleared that items exported less than 200(Tone) is 4000+, items exported greater than 800 Tone is nearly 200+

# Print all the Country Where India exported their Items

```python
#program to print all Country where India Exported Their Items
g=df.groupby('country')
for country,country_df in g:
    print(country)
```

# What are the items that are Exported From INDIA

```python
items=df.groupby("Commodity")
for item,item_df in items:
    print(item)
```

```python
years=df.groupby('year')
x=years.max()
print(x)
```

So From above analysis it is cleared that INDIA maximum exported ZINC AND ARTICLES.... in country in ZIMBABWE in every Year.
